```python
def cleanAndTokenize(text):
    narrLower=text.lower()
    narrReplaced = narrLower
    for mark in ['-','-','*']:
        narrReplaced=narrReplaced.replace(mark,' ')
    tokens=nltk.word_tokenize(narrReplaced)
    return tokens
```

```python
allTokens=[cleanAndTokenize(text) for text in allNarrators]
```

## Sentence tokenizing

```python
print(nltk.sent_tokenize(bet)[16])
# It seems that this tokenize doesn't work well because there are more than one sentences.
```

```python
betSents=nltk.sent_tokenize(bet)
```

```python
len(betSents)
```

```python
def lengthOfSentence(sent):
    tokens=nltk.sent_tokenize(sent)
    return len(tokens)
```

```python
def averageSentenceLength(narr):
    sentence=nltk.sent_tokenize(narr)
    return pd.Series([lengthOfSentence(sent) for sent in sentence]).mean()
```

```python
averageSentenceLength(bet)
```

```python
[averageSentenceLength(narr) for narr in allNarrators]
# Bruff is a lawyer so his sentence is longer.
```
